Defined a Function to clean up the N/A value in the emp_title column 

```python
def clean_title(title):
  if isinstance(title,str):
    return title.strip().title()
  else:
      return 'Unknown'
for example in examples:
  print(clean_title(example))
```

Used The Apply Function to Clean up emp_title column with the clean_title function  

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

#Define a function called term to float to convert the tem colmun to an interger value  

```python
def term_to_float(value):
  return int(value.strip().strip('months'))

df['term'] = df['term'].apply(term_to_float)

df['term'].head()
```

#defined a function to change the loan_status column value to a 1 if loan status was Current or Fully Paid and 0 if not 

```python
def loan_to_float(values):
  if values == 'Current' or values == 'Fully Paid':
    return 1
  else:
    return 0
```

#Used Function to change the entire loan_status column value to 1 or 0 

```python
df['loan_status'] = df['loan_status'].apply(loan_to_float)
df.head()
```

#Used the pandas to_datetime function to change the last_pymt_d column to a date and time then used to dt.year and the dt.month to create a month and year column from the month and year values in the last_pymt_d column 

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format= True)
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month

df.head()
```

#Defined a function to change the int_rate column to a float value while removing the leading space and the % sign from the cell value  

```python
def int_to_float(cell_value):
  return float(cell_value.strip().strip('%'))
```

Used the apply function to use int_to_float function to change the entire int_rate column to a float value

```python
df['int_rate'] = df['int_rate'].apply(int_to_float)
```

#Defined a function to convert the string into a float and strip the % sign

# Defined a variable to list the top 20 job title then defined a function to look to see if the cell value in the top if yes return the job title if no return 'Other'



```python
top_20 = df['emp_title'].value_counts()[0:20]
def job_to_other(cell_value):
  if cell_value in top_20:
    return cell_value
  else:
   return 'Other'
```
